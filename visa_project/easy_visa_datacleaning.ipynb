**Task 1.3: Inspect the dataset for missing values and handle them appropriately.**

```python
# checking the columns in the dataset
df_columns = df.columns
df_columns
```

```python
# checking the shape of the dataset
df.shape
```

```python
  # creating a copy of the data
df_visa =df.copy()
df_visa.head()
```

```python
# checking the for missing values and handling missing values
def missing_values():
    # checking the columns in the dataset
    df_visa_cols = df_visa.columns
    for col in df_visa_cols:
        missing_value_count = df[col].isna().sum()
        percent_missing = (missing_value_count/len(df))*100
        # print(f"{col} : {missing_value_count}")
        if  percent_missing >1<= 5:
            df_visa.dropna()
            print("Dropped missing values")
        elif percent_missing >5:
            num_cols = [x for x in df_visa.select_dtypes(include=["int64", "float64"]).columns] 
            cat_cols = [x for x in df_visa.select_dtypes(include=["object"]).columns]
            if col in num_cols:
                median_value= df_visa[col].median()
                df_visa[col].fillna(median_value, inplace=True)
                print("Filled with median value")
            if col in cat_cols:
                mode_value  = df_visa[col].mode()
                df[col].fillna(mode_value, inplace = True)  
                print("Filled with mode value") 
                       
        elif percent_missing == 0: 
            print("No missing values")
            print(f"{col} : {missing_value_count}")  
        else:
            print(f"{col} : {missing_value_count}")    
missing_values()  
```
